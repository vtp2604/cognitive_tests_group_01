```python
from IPython.display import display, Image, clear_output, HTML, Image
import time
import random

print("Please read: ")
print("")
print("We wish to record your response data")
print("to an anonymised public data repository.")
print("Your data will be used for educational teaching purposes ")
print("practising data and visualisation.")
print("")
print("Please type y in the box below if you consent to the upload.")
result = input(">")
if result == "y":
    print("Thanks - your data will be uploaded.")
else:
    print("No problem, we hope you enjoyed the test!")
print("")



print("Please use an anonymised ID code.")
print("To do this use:")
print("   - Two initals of childhood friend")
print("   - Two initals of favourite character")
print("Example: Bob Smith and Jack Sparrow, would give BSJS")
print("")
input_name = input("Please input your anonymised ID code")
print("")


print("Welcome to a mathematical ability test")
time.sleep(2)
print("Part of each questions will be shown for 2 seconds and disappear")
time.sleep(2)
print("Are you ready? (Press enter to start)")
input("")

count = 0
time_taken_t = 0

#Easy Difficulty, 5 questions, addition with 2 digit numbers below 25
def d1_questions():

    ''' This function is for Easy difficulty, it asks the user 5 questions, only addition between 2 digit numbers below 25'''
    
    global count
    time_taken1 = 0
    
    print("This is the Easy section, worth 1 point for each correct answer")
    time.sleep(1)
    print("Lets start")
    time.sleep(1)
    clear_output(wait=True)
    
    start_time = time.time()

    for i in range(5):
        n1 = random.randrange(1,25)
        n2 = random.randrange(1,25)
        print(n1)
        clear_output(wait=True)
        time.sleep(1)
        print("+",n2)
        time.sleep(1)
        answer = int(input("What is the answer?"))
        if answer == (n1+n2):
            print("Correct! Next question")
            clear_output(wait=True)
            time.sleep(1)
            count = count + 1
        else:
            print("Incorrect")
        clear_output(wait=True)
        time.sleep(1)
        

    end_time = time.time()

    time_taken1 = end_time - start_time
    rounded_time_taken1 = round(time_taken1,4)
    print("Your time taken for this section is: ",rounded_time_taken1," seconds")
    time.sleep(1.5)

    return time_taken1
    

#Medium difficulty, 5 questions, addition and subtraction with 2 digit numbers below 100

def d2_questions():

    ''' This function is for Medium difficulty, it asks the user 5 questions, with addition or substraction between 2 digit numbers below 100'''
    
    global count
    time_taken2 = 0
    
    print("This is the Medium section, worth 2 points for each correct answer")
    time.sleep(1)
    print("Lets start")
    time.sleep(1)
    clear_output(wait=True)

    
    start_time = time.time()

    d2_options = ['+','-']

    for i in range(5):
        n1 = random.randrange(25,100)
        n2 = random.randrange(25,100)
        d2_options = ['+','-']
        random_choice = random.choice(d2_options) 
        
        if random_choice == '+':
            x = n1 + n2
        else:
            x =  n1 - n2
        
        print(n1)
        clear_output(wait=True)
        time.sleep(1.5)
        print(random_choice,n2)
        time.sleep(1.5)
        answer = int(input("Answer?"))
        if answer == x:
            print("Correct")
            clear_output(wait=True)
            time.sleep(1)
            count = count + 2
        else:
            print("Incorrect")
            print("The Correct Answer is:",x)
            time.sleep(0.5)
        clear_output(wait=True)
        time.sleep(1)
        

    end_time = time.time()

    time_taken2 = end_time - start_time
    rounded_time_taken2 = round(time_taken2,4)
    print("Your time taken for this section is: ",rounded_time_taken2," seconds")
    time.sleep(1.5)

    return time_taken2


#Hard difficulty, 5 questions, addition and subtraction with 3 digit numbers below 500

def d3_questions():

    ''' This function is for Hard difficulty, it asks the user 5 questions, with addition or subtraction between 3 digit numbers below 500'''
    
    global count
    time_taken3 = 0
    
    print("This is the Hard section, worth 3 points for each correct answer")
    time.sleep(1)
    print("Lets start")
    time.sleep(1)
    clear_output(wait=True)

    
    start_time = time.time()

    d3_options = ['+','-']

    for i in range(5):
        n1 = random.randrange(99,500)
        n2 = random.randrange(99,500)
        d2_options = ['+','-']
        random_choice = random.choice(d3_options) 
        
        if random_choice == '+':
            x = n1 + n2
        else:
            x =  n1 - n2
        
        print(n1)
        clear_output(wait=True)
        time.sleep(1.5)
        print(random_choice,n2)
        time.sleep(1.5)
        answer = int(input("Answer?"))
        if answer == x:
            print("Correct! Next question")
            clear_output(wait=True)
            time.sleep(1)
            count = count + 3
        else:
            print("Incorrect")
            print("The Correct Answer is:",x)
            time.sleep(0.5)
        clear_output(wait=True)
        time.sleep(1)
        

    end_time = time.time()

    time_taken3 = end_time - start_time
    rounded_time_taken3 = round(time_taken3,4)
    print("Your time taken for this section is: ",rounded_time_taken3," seconds")
    time.sleep(1.5)

    return time_taken3



def maths_test():

    ''' This function runs all difficulty sections, and gives the user their score and final results'''
    
    global time_taken_t
    
    time_taken1 = d1_questions()
    time_taken2 = d2_questions()
    time_taken3 = d3_questions()
        
    print("You've completed the test!")
    time.sleep(0.5)
    time_taken_t = time_taken1 + time_taken2 + time_taken3
    print("Your total time taken is:", time_taken_t," seconds")
    time.sleep(0.5)
    print("Your total score is:",count,"out of 30")
    return time_taken_t

maths_test()

ans1 = input_name
ans2 = count
ans3 = time_taken_t

import requests
from bs4 import BeautifulSoup
import json

def send_to_google_form(data_dict, form_url):
    ''' Helper function to upload information to a corresponding google form 
        You are not expected to follow the code within this function!
    '''
    form_id = form_url[34:90]
    view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
    post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'

    page = requests.get(view_form_url)
    content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
    content = content.text[27:-1]
    result = json.loads(content)[1][1]
    form_dict = {}
    
    loaded_all = True
    for item in result:
        if item[1] not in data_dict:
            print(f"Form item {item[1]} not found. Data not uploaded.")
            loaded_all = False
            return False
        form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
    post_result = requests.post(post_form_url, data=form_dict)
    return post_result.ok
  
data_dict = {
    'Name': ans1,
    'Score': ans2,
    'Time Taken': ans3
}

form_url = "https://docs.google.com/forms/d/e/1FAIpQLSenEavD4NP2dE-0mQ7VTLTKFGkKYqh4a1UT9xjdeiXCro5e-Q/viewform"
send_to_google_form(data_dict, form_url)
```
