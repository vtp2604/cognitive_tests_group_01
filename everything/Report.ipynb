# Investigating the association of Cognitive skills and drawing comparisons between male and female performance
## Introduction

Background: 

Cognition involves the process of acquiring new knowledge and perceiving and understanding it through thoughts, experiences and the 5 senses. Complex intellectual functions involve cognition, including memory, reasoning, visuospatial function, judgement and decision-making. ([Dhakal, A. and Bobrin, B. (2023)](#dhakal_2023))
([Veron, D. (2021)](#veron_2021))

The Approximate Number System (ANS) enables the estimation of numerosities without involving words or language. It is responsible for representing and processing numerical magnitude information, exploring our intuitive, rather than deliberative understanding of num. ([Odic, D. and Starr, A. (2018)](#odic_2018)) Studies have reported a distinct correlation between ANS and mathematical abilities, where a developed ANS cognitive system might have long-term benefits on simple arithmetic skills ([Cai, D., Zhang, L., (Li, Y., Wei, W. and Geogiou, G. (2018)](#cai_2018)).Symbolic estimation is a key factor in the ANS test, which might aid children in understanding symbolic arithmetic better. Memory comes into play while recalling symbolic information and the respective magnitudes.([Lackey, D. (2017)](#lackey_2017))

Mathematical ability is about visualising a problem in your head and using traditional methods to solve the problem. Several ways to tackle mental maths problems involve figuring out the best method for you ([National Research Council (NRC) (2006)](#NRC_2006)). Spatial ability is one of the main educational abilities. While the numerical and verbal abilities have been emphasised in the classroom for children, spatial reasoning has not been. Many researchers have proved that the ability to envision an imaginary object’s shape or movement is significant in building their understanding of cognitive mapping, for instance, structures of molecules in organic chemistry ([Newcombe, N. (2010)](#newcombe_2010)). Memory is one component of the cognitive process that follows the order of perceiving, storing, and recalling information. Some numerical factors can affect the ability to memorise, for instance, age, stress, degree of attention, hours of sleep, and gender. Memory tests are crucial assessments when addressing brain health or studying psychology and neuroscience.

## Method

Here, we exhibit 4, 3-minute-tests to understand our complex intellectual function where participants are asked to consent to collecting information, including an anonymous username of a 4-character unique identifier code chosen by each user, gender, time taken, and test results, for an investigation into the relationship between cognitive tests. Upon completion, a function sends this data and question responses to a Google form.

**ANS Test:**  
A set of 64 images was created and stored in a folder to access each image using a file path, displaying 2 contrasting arrays of dots using a display image function with a loop. Blue and yellow were the chosen colours for the homogeneous dots. Arrays ranged in numerosity from 9 to 21, with various grouping ratios, manipulated to bring in variation, shown in Table 1. Each image was shown for 750 ms, and the user was asked to judge which side had more dots by clicking either the “Left” or “Right” buttons on the screen. A function created the buttons with a display timer, giving the user 3 seconds to respond. Failing to click a button within 3s would automatically display the subsequent image, recording the rses as Unanswereded\``. There was no trial-by-trial feedback.

| 9:8   | 18:16                            |
| 10:9  | 10:9, 20:18                      |

**Spatial Reasoning Test:**  
The test consists of 10 questions given to the participants. Each question asks the user to select the “correct” 2-dimensional representation when looking at a certain angle toward the 3D image displayed in the question. Only 10 questions are given because the test is aimed to be finished in 3 minutes, which makes the result more credible in terms of capturing the participant’s own cognitive thinking. Redoing the test is possible as the questions are designed to generate the 3D images randomly.

**Maths Test:**  
To test a person’s mathematical ability, the maths test was divided into 3 sections: Easy, Medium and Hard. The test involved using addition and subtraction on numbers from 1 digit to 3 digits. Without using a calculator. The numbers in each section were randomised. For example, in the Easy section, the random library was imported to generate random numbers between 1 to 25. The range of random numbers increased to 25 to 100 in the Medium section, and numbers between 100 to 500 in the Hard section. Only addition was involved in the Easy section while in the Medium and Hard section, a random choice was made between addition and subtraction to be the mathematical operator for the question.

**Memory Test:**  
The memory test involves memorising and distinguishing images, colours, and positions within three grids of increasing difficulty. For the three grids (2x2, 3x3, 4x4), participants have 10, 25, and 40 seconds to memorise the grids. The total score set for the memory test is 15, with one mark awarded for each question. If the input answer corresponds to the set answer, there will be no change to the score. However, if the input answer differs from the set answer, a deduction of one mark will be made.

## Results

**Hypothesis A - Is ANS related to maths, spatial reasoning, or memory skills?**

**Null Hypothesis:**
An individual’s Approximate Number System (ANS) skills have no association with mathematical, spatial reasoning, or memory skills. Furthermore, there are no significant differences between male and female scores on each test.

**Alternative Hypothesis:**
A statistically significant relationship exists between the 4 cognitive skills. A distinct positive correlation is predicted between ANS test performance and Maths, Spatial Reasoning, and Memory test performance. This is due to the ANS test targeting specific cognitive abilities in the brain that are each targeted by the other 3 tests.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr

# Import data from CSV file
data = pd.read_csv('ans_vs_spatial.csv')

# Extract x and y values from the data
x = data['ans']
y = data['spatial']

# Create scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('ANS result as a percentage value')
plt.ylabel('spatial reasoning result as a percentage value')
plt.title('Scatter Plot')


# Calculate the line of best fit
m, b = np.polyfit(x, y, 1)
line = m * x + b

# Add the line of best fit to the plot
plt.plot(x, line, color='red')

# Add labels and title
plt.xlabel('ans score in percentage')
plt.ylabel('spatial score in percentage')
plt.title('Scatter Plot of ANS vs Spatial reasoning test')


# Show the plot
plt.show()


data = pd.read_csv('ans_vs_spatial.csv')

# Extract the columns from the data
x = data['ans']
y = data['spatial']

# Calculate the Spearman's rank correlation coefficient
correlation, p_value = spearmanr(x, y)

# Print the correlation coefficient
print("Spearman's rank correlation coefficient:", correlation)
print("p-value:", p_value)



```

## Discussion

The histogram results are not normally distributed as expected; instead, they show an irregular figure. Which might be due to the distractions and attitude when completing the assessment. According to the scatter plot and Spearman Correlation test, the finding may imply that the cognitive function linked with the ANS activation does not impact memory spatial and calculation processing.  However, other factors still need to be considered, as recent studies have found that emotion associated with heart rate and blood pressure during ANS activation is related to enhancing attention and information retention. Both items are known to be the key factor in better spatial and memory performance.

There was no clear indication of which gender performed better in all tests. Females did comparatively better than males in the ANS test with a higher IQR. Although the median is slightly lower, more female users within that central higher range exist. There was no difference in gender from t-tests and dot plots. 

Not all the tests have considered reaction time (RT), which could not give a strong understanding of ANS acuity. Analysing accuracy in the absence of RT (or vice versa) might lead to erroneous data, where the user might have a low score due to fast decisions rather than accuracy. Besides that, all tests had an unequal number of male and female participants, which made comparisons between the genders difficult. Standardising this is essential for future reference.



## Outlook

The ANS test found that users were more accurate when the spatial arrangement was more contrasting. In addition, the colours could also have affected choices, where blue, a darker and bolder colour, might have been the preferred choice when dot ratios were similar. Using the same neutral colour next time would be interesting. The Maths test could explore a range of complicated arithmetics progressing by including calculus. Meanwhile, by matching similar shapes, the Memory test could be more interactive, which may increase the degree of attention and gain more precise results.
